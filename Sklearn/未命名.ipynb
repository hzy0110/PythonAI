```python
from sklearn.preprocessing import MinMaxScaler
minMax = MinMaxScaler()
data = np.array([1,2,3])
dftest = pd.DataFrame(data)
dftest_mm = minMax.fit_transform(dftest)
print(dftest_mm)
```

```python
from sklearn import preprocessing
# 标准化 (X-X_mean)/std
X_scaled = preprocessing.scale(data)
print(X_scaled)
print(np.var(X_scaled))
```

```python
print(np.std(data))
print(np.var(data)) # 2/3 = 0.666
```

```python
# data = np.array([1,2,3])
print(type(data))
#  ddof : int, optional  控制自由度，自由度=N - ddof，N是数据的长度。当ddof=1时，就是样本标准差，当ddof=0时，是总体标准差
print(np.std(data, ddof = 1)) 
print(np.std(data, ddof = 0)) 
print(np.mean(data)) # 均值
print('sum',((data - np.mean(data)) ** 2).sum() )
print('data.size',data.size)
print(np.sqrt(0.66666))
print(np.sqrt(((data - np.mean(data)) ** 2).sum() / (data.size - 1)))
print(np.sqrt(((data - np.mean(data)) ** 2).sum() / (data.size - 0)))
print(np.sqrt(( data.var() * data.size) / (data.size - 1)))
```
